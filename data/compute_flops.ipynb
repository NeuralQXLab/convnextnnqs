To compute FLOPs of a single evaluation of specified network

```python
import netket as nk
from deepnets.net import ConvNext
from deepnets.system import Shastry_Sutherland
import flax.linen as nn
import jax
import jax.numpy as jnp
```

```python
#Initialize the desired network
L = 10
J = [0.8,1]
system = Shastry_Sutherland(L,J)
n_blocks = (2,)
features = (72,)
expansion_factor = 2
net_type = "Vanilla"
kernel_width = 2
downsample_factor = 2
final_features = features[0]
init_kernel_width = 1
output_depth = 1
net = ConvNext(n_blocks,features,expansion_factor,net_type,kernel_width,downsample_factor,final_features,init_kernel_width,output_depth,system).network
params = net.init(jax.random.PRNGKey(0),jnp.ones(L**2))
```

```python
#Tabulate info, including FLOPs
tab_fn = nn.tabulate(net,jax.random.PRNGKey(0),compute_flops=True)
print(tab_fn(jnp.ones(L**2)))
```
